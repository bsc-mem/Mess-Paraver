```python
def get_trace_df(trace_file, verbose=False):        
    df = pd.DataFrame(columns=['timestamp', 'wr', 'rr', 'bw', 'max_bw', 'lat', 'min_lat', 'max_lat'])
    with open(trace_file) as f:
        # all_lines = f.readlines()
        first_line = True
        for i, line in enumerate(f):
            if first_line:
                first_line = False
                continue
            if verbose:
                if i%1000 == 0:
                    print(i)

            sp = line.split(':')
            row = defaultdict()
            # timestamp is at 5th place
            row['timestamp'] = int(sp[5])
            # process subsequent metric IDs and values after the timestamp
#             for metric_id, val in zip(sp[6:-1], sp[7:]):
            for i in range(6, len(sp)-1, 2):
                metric_id = sp[i]
                val = sp[i+1].strip()
                if row['timestamp'] == 81702482874:
                    print(metric_id, val)
                val = int(val)
                if metric_id == '1':
                    row['wr'] = val
                    row['rr'] = 100 - val
                elif metric_id == '2':
                    row['bw'] = val
                elif metric_id == '3':
                    row['max_bw'] = val
                elif metric_id == '4':
                    row['lat'] = val
                elif metric_id == '5':
                    row['min_lat'] = val
                elif metric_id == '6':
                    row['max_lat'] = val
            df = df.append(row, ignore_index=True)
        # perform a forward fill for copying above values of NaN (prvparser is purpously omitting equal consecutive values of a metric)
        df = df.ffill()
        return df
```
